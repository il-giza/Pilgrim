```python

s = {}
i = 0
s[i] = PilgrimCross(0)
#s[i].show_info()

paths = {} 

#for t in tqdm(range(6000000)):

for t in range(6000000):    
    try:
        j=i+1
        s[j] = s[i].get_next_cross()
#        s[j].show_info()
        if s[j].finish:
            print('-' * 30 +  ' Yes! ' + '-' * 30)
#            s[j].show_info()
            paths[t] = s[j].show_path()
            print(t, paths[t])
#            print('-' * 46)
        i += 1 
    except Exception:
        del s[i]
        i -= 1 
    if i < 0:
        print('No more next step. t=', t, 'len(s)=', len(s))
        break
        
```

```python

d = 0.1 

for i in paths.keys():
#    print('-'*20)
    x = []
    y = []
    b = []
    
    for j in paths[i][::-1]:
#        print(i, j) 
        x.append(j[2][0])
        y.append(j[2][1])
        b.append(j[3])
        
    x_ = [(x[i+1]+x[i])/2  for i in range(len(x)-1)]
    y_ = [(y[i+1]+y[i])/2  for i in range(len(y)-1)]
        
    plt.figure(figsize=(10,10))    
    plt.gca().invert_yaxis()    
    plt.plot(x, y, ':o')    
    
    for t in range(len(x_)):
        plt.text(x_[t], y_[t], str(t) + '( ' + str(b[t+1]) + ' )')
    
    
    for sx in range(1,5):
        for sy in range(1,5):
            ax = sx+d
            bx = sx+1-d
            ay = sy+d
            by = sy+1-d
            plt.plot([ax, ax, bx, bx, ax],
                     [ay, by, by, ay, ay],
                    ) 
    plt.axis('off')
    plt.show();     


```
